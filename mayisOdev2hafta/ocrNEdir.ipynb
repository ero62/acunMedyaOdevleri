    optik karekter tanıma yada kısaca OCr bilgisayar ortamında bulunmayan yazılı dokümanların özel tarayıcılar veya normal olarak taranmış resimlerinin FineRead ve OmniPage gibi bazı özel programlar arayıcılığıyla bilgisayar ortamın düzenlenebilecek sayısal halde aktarılmasıdır. ## OCR türleri 
    4 ceşit ocr programı vardır ve giderek dahada karmaşık hale gelirler
    - Basit ocr : Analiz, taranan karakterleri depolanan gliflerle karşılaştıran karakter karakter desen eşleştirmesidir. Çok sayıda potansiyel yazı tipi ve dil kombinasyonuyla, analiz edilebilecek belge türleri sınırlıdır.
    - Optik işaret tanıma : nketlerdeki baloncuklar veya bir formdaki imza gibi işaretli kutuları ve  diğer işaretleri , ayrıca logoları, sembolleri ve filigranları tanımlamak için. Hepsi, basit OCR'de olduğu gibi, depolanan görüntülerle eşleştirilerek tanımlanabilir.
    - akıllı karekter tanıma : Daha önce belirtildiği gibi, ICR yapay zekanın gücünü devreye sokar.  ML  veya  derin öğrenmeyi kullanarak , OCR programı tıpkı insanlar gibi okumayı öğrenir: sürekli pratik ve eğitim yoluyla. Bir sinir ağı, ayırt edici nitelikleri aramak için metni tekrar tekrar inceler: eğrilerin, kesişimlerin, çizgilerin ve döngülerin konumları.
    - Akıllı kelime tanıma : Bu, önceki ICR tanımanın doğal evrimidir, ancak artık yapay zeka, tek bir resimdeki bir kelimeyi tanıyacak şekilde eğitildi ve bu da onu daha hızlı hale getirdi.

```python
from ultralytics import YOLO
import cv2
import pytesseract
import os
import pandas as pd

pytesseract.pytesseract.tesseract_cmd = r"C:\\Program Files\\Tesseract-OCR\\tesseract.exe"


model = YOLO(r"C:\\Users\\admin\\Desktop\\Projelerim\\python\\AcunMedya_YZ\\Conda_YZ\\cv\\runs\\detect\\train2\\weights\\best.pt")


image_dir = r"C:\\Users\\admin\\Desktop\\Projelerim\\python\\AcunMedya_YZ\\Conda_YZ\\cv\\dataset\\images"
output_data = []
output_image_dir = "islenmis_gorseller"
os.makedirs(output_image_dir, exist_ok=True)


for file in os.listdir(image_dir):
    if file.lower().endswith((".jpg", ".jpeg", ".png")):
        image_path = os.path.join(image_dir, file)
        img = cv2.imread(image_path)

        results = model(image_path)[0]

        for box in results.boxes.xyxy:
            x1, y1, x2, y2 = map(int, box)

            
            cropped = img[y1:y2, x1:x2]

          
            gray = cv2.cvtColor(cropped, cv2.COLOR_BGR2GRAY)
            text = pytesseract.image_to_string(gray, config='--psm 7')

            
            plate_text = text.strip().replace('\n', '').replace(' ', '')

            
            cv2.putText(img, plate_text, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 255), 2)
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)

            
            output_data.append({"dosya": file, "plaka": plate_text})

       
        output_path = os.path.join(output_image_dir, file)
        cv2.imwrite(output_path, img)


df = pd.DataFrame(output_data)
df.to_csv("plaka_sonuclar.csv", index=False)

print("✅ OCR işlemi tamamlandı ve sonuçlar:")
print("- Görseller 'islenmis_gorseller' klasörüne kaydedildi.")
print("- Plaka metinleri 'plaka_sonuclar.csv' dosyasına yazıldı.")
```

```python
from ultralytics import YOLO
import cv2
import pytesseract
import os
import pandas as pd


pytesseract.pytesseract.tesseract_cmd = r"C:\\Program Files\\Tesseract-OCR\\tesseract.exe"


model = YOLO(r"C:\\Users\\admin\\Desktop\\Projelerim\\python\\AcunMedya_YZ\\Conda_YZ\\cv\\runs\\detect\\train2\\weights\\best.pt")


image_dir = r"C:\\Users\\admin\\Desktop\\Projelerim\\python\\AcunMedya_YZ\\Conda_YZ\\cv\\dataset\\images"
output_image_dir = "islenmis_gorseller"
os.makedirs(output_image_dir, exist_ok=True)

output_data = []


for file in os.listdir(image_dir):
    if file.lower().endswith((".jpg", ".jpeg", ".png")):
        image_path = os.path.join(image_dir, file)
        img = cv2.imread(image_path)

        results = model(image_path)[0]

        for box in results.boxes.xyxy:
            x1, y1, x2, y2 = map(int, box)
            cropped = img[y1:y2, x1:x2]

        
            gray = cv2.cvtColor(cropped, cv2.COLOR_BGR2GRAY)
            gray = gray[:, 30:]  
            _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
            text = pytesseract.image_to_string(thresh, config='--psm 8')
            plate_text = text.strip().replace('\n', '').replace(' ', '')
           

            
            cv2.putText(img, plate_text, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 255), 2)
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)

            
            output_data.append({"dosya": file, "plaka": plate_text})

        
        output_path = os.path.join(output_image_dir, file)
        cv2.imwrite(output_path, img)

df = pd.DataFrame(output_data)
df.to_csv("plaka_sonuclarV2.csv", index=False)

print("✅ Tamamlandı:")
print("- OCR sonuçları 'plaka_sonuclar.csv' dosyasına yazıldı.")
print("- Görseller 'islenmis_gorseller' klasörüne kaydedildi.")
```
