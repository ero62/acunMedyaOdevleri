# davis boulding methodu nedir 
    Bir kümleleme değerlendirme metriğidir. Bu metrik, denetimsiz öğrenme algoritmalarının performansını, kümelerin içsel özelliklerine dayanarak değerlendiri.
    Her bir kümenin kendi içindeki sıklığı ve diğer kümelerden ne kadar ayrıştığını ölçerek, kümeleme kalitesini değerlendirir . Daha düşük DBI değerleri, daha iyi kümeleme sonuçlarını gösterir çünkü bu, kümelerin hem kendi içinde daha sıkı hem de diğer kümelerden daha iyi ayrıldığını belirtir. 
## avantajları 
    etiket gerektirmez
    yorumlaması kolay 
    esneklik : farklı kümeleme algoritmaları ve mesafe ölçümleriyle kullanılabilir 
## dezavantajları 
    aykırı değerlere ve gürültüye duyarlılık 
    küme şekli ve boyutu varsayımları 
    ölçekleme ve mesafe ölçümüne bağımlılık 
    yüksek boyutlu verilerde yorumlama zorluğu 



```python
# Gerekli kütüphanelerin içe aktarılması
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
X, y_true = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)


db_score = davies_bouldin_score(X, labels)
print(f"Davies-Bouldin Skoru: {db_score:.4f}")


plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Merkezler')
plt.title('KMeans Kümeleme Sonucu')
plt.legend()
plt.show()
```
